```python
#Grade Assignment Problem:
#Assign grades based on marks: Marks >= 90: Grade A Marks >= 75: Grade B Marks >= 50: Grade C Otherwise: Grade F:
```

```python
marks = int(input("Enter marks: "))

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Grade F")
```

```python
#Calculate electricity bill based on usage:
#Units <= 100: ₹5/unit Units <= 200: ₹10/unit Units > 200: ₹15/unit:
```

```python
units = int(input("Enter units: "))

if units <= 100:
    bill = units * 5
elif units <= 200:
    bill = units * 10
else:
    bill = units * 15

print("Bill Amount:", bill)
```

```python
#Classify a triangle as equilateral, isosceles, or scalene based on side lengths.
```

```python
a = int(input("Side a: "))
b = int(input("Side b: "))
c = int(input("Side c: "))

if a == b == c:
    print("Equilateral")
elif a == b or b == c or a == c:
    print("Isosceles")
else:
    print("Scalene")
```

```python
#Classify BMI:
#BMI < 18.5: Underweight 18.5 <= BMI < 24.9: Normal Weight 25 <= BMI < 29.9: Overweight BMI >= 30: Obese
```

```python
#Check if three sides form a valid triangle. A triangle is valid if the sum of any two sides is greater than the third side.
```

```python
a = int(input("Side a: "))
b = int(input("Side b: "))
c = int(input("Side c: "))

if a + b > c and a + c > b and b + c > a:
    print("Valid Triangle")
else:
    print("Invalid Triangle")
```

```python
#Check if a given year is a leap year. A year is a leap year if:
#Divisible by 4 Not divisible by 100, unless also divisible by 400
```

```python
year = int(input("Enter year: "))

if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
    print("Leap Year")
else:
    print("Not a Leap Year")
```
